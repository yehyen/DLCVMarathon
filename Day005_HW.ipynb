```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

import json
import requests

r = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=true")
rf = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=false")
data = json.loads(r.text)
dataf = json.loads(r.text)
sumPop = 0
sumNonpop = 0
sumPerfect = 0

print(len(data), data, sep="\n", end="\n\n")
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for d in data:
    print(d["title"], d["createdAt"], d["commentCount"], d["likeCount"], end="\n")
    sumPop += d["commentCount"]
    sumPerfect += d["likeCount"]
print(round(sumPop / len(d), 2), end='\n') 
print(round(sumPerfect / len(d), 2), end='\n') 

# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
for df in dataf:
    sumNonpop += df["commentCount"]
    sumPerfect += df["likeCount"]
print(round(sumNonpop / len(df), 2), end='\n') 
print(round(sumPerfect / len(df), 2), end='\n') 
```
