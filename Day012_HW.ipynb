## 作業目標

* ① 取出今天所有的發文

* ③ 哪一個類別的文章最多




### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.ettoday.net/news/news-list.htm"

r = requests.get(url, "html5lib")
soup = BeautifulSoup(r.text, "html5lib")

# Q1. 取出今天所有的發文
for d in soup.find(class_="part_list_2").find_all("h3"):
    print(d.find(class_="date").text, d.find_all("a")[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
# Q2. 如果想要依類別分類，怎麼儲存會比較好?
res = {}
for i in soup.find(class_="part_list_2").find_all("h3"):
    date = i.find(class_="date").text
    title = i.find_all("a")[-1].text
    tag = i.select(".tag")[0].text

    # 將dict的key初始化為tag資料, 值為[]
    res.setdefault(tag, [])
    # res[tag]的值從後面加入
    res[tag].append({"title":title, "date":date})
print(res)
```

### ③ 哪一個類別的文章最多

```python
# Q.3 哪一個類別文章最多
data = []
for j in res:
    print(j, len(res[j]))

    # 存成dict
    d = {}
    d["tag"] = j
    d["count"] = len(res[j])
    # print(d)

    # 為了之後好比較和排序，必須把每組加到同一個list
    data.append(d)

# 存成新dict，要依照key為count時數值排序，並將順序反向[::1]
for k in sorted(data, key=lambda k: k["count"])[::-1]:
    print(k)
```
