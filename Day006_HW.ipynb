## 作業目標

    1. 取出知乎問題發問時間
    2. 取出第一筆與最後一筆回答的時間


### 1. 取出知乎問題發問時間

```python
# 1. 取出知乎問題發問時間

import json
import requests
from time import *


headers = {"user-agent":"my-app/0.01"}
r = requests.get("https://www.zhihu.com/api/v4/questions/55493026/answers", headers=headers)

data = json.loads(r.text)
latest = 0
for d in data["data"]:
    created = d["question"]["created"]
    timef = strftime("%m/%d/%y, %H:%M:%S", localtime(created))
  
    if d["created_time"] > latest:  
        lastestAnswer = d["created_time"]
print(timef)
```

### 2. 取出第一筆與最後一筆回答的時間

```python
# 2. 取出第一筆與最後一筆回答的時間

firstAnswer = data["data"][0]["created_time"]
print(strftime("%m/%d/%y, %H:%M:%S", localtime(firstAnswer)))

print(strftime("%m/%d/%y, %H:%M:%S", localtime(lastestAnswer)))
```
